A mensagem &quot;Remove columns with datetime64 data type before applying sum&quot; significa que, ao tentar aplicar a soma (`sum`) em um DataFrame do Pandas, você deve remover ou ignorar colunas que contêm dados do tipodatetime64`. Isso ocorre porque a operação de soma não faz sentido para dados de data e hora.

 Por que isso acontece?

1. **Tipo de Dados Inível**: O comando `sum()` é usado para somar valores numéricos. Colunas com tipo de dado `datetime64` não são numéricas, não podem ser somadas diretamente.
  
2. **Resultado Inválido**: Se você tentar som valores de data e hora, o Pandas não saberá como lidar com isso e pode gerar um erro ou resultados inesperados.
